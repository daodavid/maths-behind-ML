```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from nose.tools import *

from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import MinMaxScaler
```

```python
wine = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv",delimiter=';')
```

```python
wine.info()
```

```python
wine.head()
```

```python
wine_features = wine.drop("quality", axis = 1)
```

```python
model =LinearRegression()
model.fit(wine_features,wine['quality'])
model.intercept_ 
```

```python
score = model.score(wine_features,wine['quality'])
score
```

```python
poly =  PolynomialFeatures() #interaction_only=True
wine_features_quad = poly.fit_transform(wine_features)
model = LinearRegression()
model.fit(wine_features_quad,wine['quality'])
model.intercept_ 
```

```python
score = model.score(wine_features_quad,wine['quality'])
score
```
